##Q1) What is a random variable in probability theory ?

Ans:In probability theory, a **random variable** is a variable that takes on different values based on the outcome of a random event or experiment. It is a function that assigns a numerical value to each possible outcome in a sample space.

There are two main types of random variables:

1. **Discrete Random Variable**: This type takes on a finite or countably infinite set of values. For example, the number of heads in a series of coin flips (which could be 0, 1, or 2 heads if flipping two coins) is a discrete random variable.

2. **Continuous Random Variable**: This type can take on an infinite number of values within a given range. For example, the height of a randomly chosen person is a continuous random variable because it could be any value within a certain range, like between 150 cm and 200 cm.

Random variables are often denoted by letters such as \(X\), \(Y\), or \(Z\). The probability distribution of a random variable describes the likelihood of different outcomes.

- For discrete random variables, this is often represented as a **probability mass function** (PMF), which gives the probability of each possible value.
- For continuous random variables, the **probability density function** (PDF) is used, and probabilities are derived by integrating the PDF over a range of values.

##Q6) What is a discrete uniform distribution ?

Ans:A **Discrete Uniform Distribution** is a type of probability distribution where all outcomes of a discrete random variable have the same probability of occurring.

### Key Characteristics:
- The random variable can take **finite** values.
- Each possible outcome has the **same probability**.
- The distribution is "uniform" because all probabilities are equal.

### Example:
For a fair six-sided die, the possible outcomes are 1, 2, 3, 4, 5, and 6. Each outcome has an equal probability of occurring:
\[
P(X = x) = \frac{1}{6} \quad \text{for} \quad x \in \{1, 2, 3, 4, 5, 6\}
\]
This is a **discrete uniform distribution** because each side of the die has the same probability of showing up.

### General Formula:
For a discrete uniform distribution where the random variable \( X \) can take \( n \) values (say \( x_1, x_2, \dots, x_n \)), the probability of each outcome is:
\[
P(X = x_i) = \frac{1}{n} \quad \text{for each } i \in \{1, 2, \dots, n\}
\]

This distribution is often used when all outcomes in a finite sample space are equally likely.

##Q7)What are the key properties of a Bernoulli distribution ?

Ans:The **Bernoulli Distribution** is a discrete probability distribution that describes a random experiment with exactly **two possible outcomes**: success (often denoted as 1) or failure (denoted as 0). It's often used to model scenarios like coin flips, yes/no questions, or pass/fail outcomes.

### Key Properties of a Bernoulli Distribution:

1. **Two Outcomes**:
   - The random variable \( X \) can take only two values: 0 (failure) and 1 (success).

2. **Parameter \( p \)**:
   - \( p \) is the probability of success (i.e., \( P(X = 1) = p \)).
   - The probability of failure is \( 1 - p \) (i.e., \( P(X = 0) = 1 - p \)).

3. **Probability Mass Function (PMF)**:
   The PMF of a Bernoulli random variable \( X \) is:
   \[
   P(X = 1) = p \quad \text{and} \quad P(X = 0) = 1 - p
   \]
   where \( 0 \leq p \leq 1 \).

4. **Mean (Expected Value)**:
   The mean or expected value \( E[X] \) of a Bernoulli distribution is:
   \[
   E[X] = p
   \]

5. **Variance**:
   The variance \( \text{Var}(X) \) of a Bernoulli distribution is:
   \[
   \text{Var}(X) = p(1 - p)
   \]

6. **Support**:
   The distribution is supported on the set \( \{0, 1\} \), meaning the random variable can only take values 0 or 1.

7. **Applications**:
   - The Bernoulli distribution is used to model binary or dichotomous events (like success/failure, pass/fail, heads/tails).
   - It's the basis for the **Binomial Distribution**, which models the number of successes in a fixed number of independent Bernoulli trials.

##Q8)What is the binomial distribution, and how is it used in probability ?

Ans:The **Binomial Distribution** is a discrete probability distribution that models the number of **successes** in a fixed number of independent **Bernoulli trials** (experiments with two outcomes: success or failure). Each trial has the same probability of success.

### Key Characteristics of the Binomial Distribution:
1. **Fixed Number of Trials**:
   - The experiment is repeated \( n \) times (e.g., flipping a coin 10 times, rolling a die 10 times).
   
2. **Two Outcomes Per Trial**:
   - Each trial results in either a success (denoted as 1) or a failure (denoted as 0).
   
3. **Constant Probability of Success**:
   - The probability of success \( p \) is the same for each trial, and the probability of failure is \( 1 - p \).
   
4. **Independent Trials**:
   - The outcome of one trial does not affect the outcomes of other trials (i.e., trials are independent).

### Binomial Distribution Formula:
The probability of observing exactly \( k \) successes (out of \( n \) trials) in a binomial experiment is given by the **Binomial Probability Mass Function (PMF)**:

\[
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
\]

Where:
- \( n \) = number of trials.
- \( k \) = number of successes.
- \( p \) = probability of success on a single trial.
- \( \binom{n}{k} = \frac{n!}{k!(n - k)!} \) is the **binomial coefficient**, representing the number of ways to choose \( k \) successes from \( n \) trials.

### Key Properties:
1. **Mean (Expected Value)**:
   \[
   E[X] = n \cdot p
   \]
   The expected number of successes is \( n \) times the probability of success.
   
2. **Variance**:
   \[
   \text{Var}(X) = n \cdot p \cdot (1 - p)
   \]
   The variance of the binomial distribution depends on the number of trials and the probability of success.

3. **Standard Deviation**:
   \[
   \text{SD}(X) = \sqrt{n \cdot p \cdot (1 - p)}
   \]

### When to Use the Binomial Distribution:
- When you have a **fixed number of trials** (\( n \)).
- Each trial has only two possible outcomes (success or failure).
- The probability of success (\( p \)) is the same for each trial.
- The trials are **independent**.

### Example Use Case:
Suppose you flip a fair coin 10 times (so \( n = 10 \), and \( p = 0.5 \)). The binomial distribution can be used to calculate the probability of getting exactly 7 heads (successes):
\[
P(X = 7) = \binom{10}{7} (0.5)^7 (0.5)^{3} = \binom{10}{7} (0.5)^{10}
\]
Where \( \binom{10}{7} \) is the number of ways to get 7 heads in 10 flips.

### Applications of the Binomial Distribution:
- **Quality control**: To find the probability of defective items in a batch.
- **Survey sampling**: To model the probability of a certain number of people answering "yes" in a survey.
- **Marketing**: To calculate the probability of a certain number of customers buying a product based on conversion rates.

##Q13)What is the Central Limit Theorem (CLT), and why is it critical in statistics)  ?

Ans:The **Central Limit Theorem (CLT)** is a fundamental concept in statistics that states:

- **Regardless of the population's distribution**, the **sampling distribution of the sample mean** will approach a **normal distribution** as the sample size increases, **provided the samples are independent and identically distributed (i.i.d.)**.
- This is true for **any population distribution** (whether normal or not), as long as the sample size is large enough (typically \( n \geq 30 \)).

### Key Points of the CLT:
1. **Sample Mean Distribution**: If you take multiple random samples from a population and calculate their means, those sample means will form a distribution. As the sample size increases, this distribution of sample means will approximate a normal distribution, even if the original population is not normally distributed.
   
2. **Normality**: The larger the sample size, the more the distribution of sample means will resemble a normal distribution, regardless of the shape of the original population.

3. **Mean of the Sample Mean**: The mean of the sampling distribution of the sample mean will be equal to the population mean (\( \mu \)).

4. **Standard Deviation (Standard Error)**: The standard deviation of the sampling distribution (called the **standard error**) will be equal to the population standard deviation (\( \sigma \)) divided by the square root of the sample size (\( \sqrt{n} \)):

   \[
   \text{Standard Error} = \frac{\sigma}{\sqrt{n}}
   \]

### Why the Central Limit Theorem is Critical:
1. **Facilitates Inference**:
   - The CLT allows statisticians to make inferences about a population using sample data. Even if the underlying population is not normally distributed, the sample mean distribution will approximate normality for sufficiently large samples, making hypothesis testing and confidence intervals possible.

2. **Foundation for Statistical Methods**:
   - Many statistical methods (e.g., z-tests, t-tests, confidence intervals) rely on the assumption of normality. Thanks to the CLT, these methods can be applied to large sample sizes regardless of the populationâ€™s distribution.

3. **Justifies Approximation to Normal Distribution**:
   - It justifies why, in many cases, statistical procedures based on the normal distribution work well even if the population is not normal, as long as the sample size is large enough.

4. **Simplifies Analysis**:
   - By ensuring that sample means approach a normal distribution, the CLT makes it easier to analyze and interpret data, even when the exact distribution of the population is unknown.

### Example:
- If you're measuring the average height of individuals in a city, the heights of the individuals may not follow a normal distribution. However, if you take many random samples of 30 people (or more) and calculate their average heights, the distribution of those sample means will approximate a normal distribution, even if the original data is not normal.

##Q16) How do you calculate a Z-score, and what does it represent ?

### **Formula for Z-Score:**

The Z-score is calculated using the following formula:

\[
Z = \frac{X - \mu}{\sigma}
\]

Where:
- \( X \) is the value or data point you are analyzing.
- \( \mu \) is the mean of the population or sample.
- \( \sigma \) is the standard deviation of the population or sample.

### **Interpretation of a Z-Score:**
- **Z = 0**: The data point is exactly at the mean.
- **Z > 0**: The data point is above the mean.
- **Z < 0**: The data point is below the mean.
- **Z = 2**: The data point is 2 standard deviations above the mean.
- **Z = -1**: The data point is 1 standard deviation below the mean.

### **What Does the Z-Score Represent?**
- The Z-score tells you **how far** a particular value is from the mean, in terms of the standard deviation.
- It allows for comparing values across different distributions, even if those distributions have different means and standard deviations.

- The interval is usually constructed using a **point estimate** along with a margin of error, which is based on sample variability and the desired confidence level (e.g., 95% confidence).

  **Example**: If you estimate that the average height of students is between 158 cm and 162 cm with 95% confidence, this range is an **interval estimate** of the population mean.

  **Key Features**:
  - It provides a **range** (upper and lower bounds) within which the true population parameter is likely to lie.
  - It includes a **level of confidence** (e.g., 95%, 99%) indicating the likelihood that the interval contains the true parameter.

##Q19)What is the relationship between a Z-score and a confidence interval ?

Ans:The **relationship between a Z-score and a confidence interval** lies in the fact that the Z-score is used to calculate the **margin of error** for constructing a **confidence interval (CI)**, particularly when the population standard deviation is known.

### Key Concepts:
1. **Z-Score**:
   - The Z-score represents the number of **standard deviations** a data point or statistic is away from the mean of a distribution.
   - For a confidence interval, the Z-score corresponds to the critical value for a specific confidence level, reflecting how many standard deviations we need to go to capture a given percentage of the data.

2. **Confidence Interval**:
   - A confidence interval provides a range within which we expect the population parameter (such as the population mean) to lie, based on sample data.
   - The **confidence level** (e.g., 95% or 99%) defines how confident we are that the interval contains the true population parameter.

### **Using Z-Score to Calculate a Confidence Interval**:

When constructing a confidence interval for a population mean using the **Z-distribution** (typically when the sample size is large or the population standard deviation is known), the formula is:

\[
\text{CI} = \bar{X} \pm Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}
\]

Where:
- \( \bar{X} \) is the sample mean.
- \( Z_{\alpha/2} \) is the **Z-score** corresponding to the confidence level (e.g., 1.96 for 95% confidence).
- \( \sigma \) is the population standard deviation.
- \( n \) is the sample size.

### How Z-Score Relates to Confidence Intervals:
- **Z-Score for Confidence Levels**: The Z-score used in this formula corresponds to the **critical value** that marks the boundary of the confidence interval. For example:
  - For a **95% confidence level**, the Z-score is approximately **1.96**, meaning we expect the true population mean to fall within 1.96 standard deviations of the sample mean, 95% of the time.
  - For a **99% confidence level**, the Z-score is approximately **2.576**, meaning we expect the true population mean to fall within 2.576 standard deviations of the sample mean, 99% of the time.

- **Margin of Error**: The term \( Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}} \) in the CI formula is the **margin of error**. It represents how much the sample mean can deviate from the true population mean. The Z-score helps determine the **width** of the confidence interval â€” a larger Z-score (for a higher confidence level) results in a wider interval, indicating more uncertainty.

### **Example**:
Letâ€™s say you have a sample mean height (\( \bar{X} \)) of 170 cm, the population standard deviation (\( \sigma \)) is 10 cm, and the sample size (\( n \)) is 100. For a 95% confidence interval:

- The Z-score for a 95% confidence level is **1.96**.
- The formula for the confidence interval is:
  \[
  \text{CI} = 170 \pm 1.96 \times \frac{10}{\sqrt{100}}
  \]
  \[
  \text{CI} = 170 \pm 1.96 \times 1
  \]
  \[
  \text{CI} = 170 \pm 1.96
  \]
  So, the **95% confidence interval** is **(168.04, 171.96)**.

This means you are 95% confident that the true population mean height lies between 168.04 cm and 171.96 cm.

##Q20) How are Z-scores used to compare different distributions ?

Ans:**Z-scores** are a powerful tool for comparing values from different distributions because they **standardize** data points, making them comparable regardless of the original units, scale, or distribution. Here's how Z-scores help in comparing different distributions:

### **How Z-Scores Standardize Data**:
- A **Z-score** tells you how many **standard deviations** a particular value is from the **mean** of the distribution.
- The formula for a Z-score is:

  \[
  Z = \frac{X - \mu}{\sigma}
  \]
  
  Where:
  - \( X \) is the data point you're comparing.
  - \( \mu \) is the mean of the distribution.
  - \( \sigma \) is the standard deviation of the distribution.

- By converting data points into Z-scores, you remove the influence of the **original scale** (e.g., centimeters, dollars, etc.), which allows you to compare values across different distributions directly.

### **Using Z-Scores to Compare Different Distributions**:
1. **Comparing Data from Different Distributions**:
   - **Different scales**: Suppose you have two distributions: one representing the heights of people (in centimeters) and another representing test scores (in points). The original units (height vs. score) make direct comparison difficult.
   - When you convert each data point from both distributions to Z-scores, you transform them into **standardized scores**, representing how far they are from their respective means in terms of standard deviations.
   - This allows for a **fair comparison**: For example, a Z-score of +2 in one distribution indicates the data point is 2 standard deviations above the mean, and a Z-score of +2 in another distribution also indicates the data point is 2 standard deviations above the mean, regardless of the original units.

2. **Comparing Different Distributions with Different Means and Standard Deviations**:
   - Different distributions may have different means and standard deviations. A Z-score standardizes for these differences, making the comparison based solely on the **relative position** within each distribution, not the absolute values.
   - For example:
     - In one distribution (e.g., heights of people), the mean is 160 cm with a standard deviation of 10 cm.
     - In another distribution (e.g., exam scores), the mean is 75 points with a standard deviation of 15 points.
     - A score of 170 cm in the first distribution would correspond to a Z-score of \( Z = \frac{170 - 160}{10} = 1 \) (1 standard deviation above the mean).
     - A score of 90 points in the second distribution would correspond to a Z-score of \( Z = \frac{90 - 75}{15} = 1 \) (also 1 standard deviation above the mean).
   - Both values (170 cm in the first and 90 points in the second) have the same relative position to their respective means, so they can be directly compared using their Z-scores.

3. **Comparing Different Distributions with Different Shapes**:
   - Z-scores are most useful when the distributions are **normal** or approximately normal. However, they can still help compare data from distributions with different shapes (skewed, bimodal, etc.), although caution is needed if the distributions are highly non-normal.
   - For non-normal distributions, comparing Z-scores can still give insights into how extreme or unusual a data point is within its own distribution, even if the overall shape of the distribution is not symmetric.

4. **Real-World Example**:
   - Imagine you want to compare the performance of two students:
     - **Student A** scored 85 on a math test with a mean score of 70 and a standard deviation of 5.
     - **Student B** scored 90 on a history test with a mean score of 75 and a standard deviation of 10.
   - To compare their performances, convert each score to a Z-score:
     - Student A's Z-score: \( Z_A = \frac{85 - 70}{5} = 3 \)
     - Student B's Z-score: \( Z_B = \frac{90 - 75}{10} = 1.5 \)
   - **Interpretation**: Student Aâ€™s score is 3 standard deviations above the mean, while Student Bâ€™s score is 1.5 standard deviations above the mean. Therefore, even though Student B scored higher, Student A performed relatively better compared to their peers.

##Q21)What are the assumptions for applying the Central Limit Theorem ?

Ans:The **Central Limit Theorem (CLT)** is a fundamental concept in statistics that allows us to approximate the distribution of sample means as a normal distribution, even if the population distribution is not normal. However, there are certain **assumptions** that must be met for the CLT to apply effectively:

### **Assumptions for the Central Limit Theorem**:

1. **Random Sampling**:
   - The data must be **randomly sampled** from the population. This ensures that each observation has an equal chance of being selected and that the sample is representative of the population.

2. **Independent Observations**:
   - The observations in the sample must be **independent** of each other. This means that the value of one observation should not influence or depend on another. This assumption is crucial to avoid biases and ensure the validity of the CLT.

3. **Sample Size**:
   - The sample size must be **sufficiently large**. While there is no strict cutoff, a general rule of thumb is that the sample size \( n \) should be at least **30** for the CLT to hold. For **small sample sizes** (less than 30), the population should be approximately **normally distributed** for the CLT to apply effectively.
   - If the sample size is small and the population distribution is highly skewed or non-normal, the sample mean may not be approximately normal, and the CLT may not apply well.

4. **Finite Variance**:
   - The population from which the samples are drawn must have a **finite variance**. This means the population should not have infinite variability or outliers that could skew the results.

5. **Sample Without Replacement (if applicable)**:
   - If the sampling is **without replacement**, the population should be **large enough** relative to the sample size (typically, the sample size should not exceed 10% of the population size). This ensures that sampling without replacement does not affect the independence of observations too much.

##Q10)Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

Ans:

```python
import numpy as np
import matplotlib.pyplot as plt

def central_limit_theorem(population_dist, sample_size, num_samples, dist_params):
    """
    Apply the Central Limit Theorem (CLT) by sampling from a non-normal population distribution.

    Parameters:
    - population_dist: The distribution function to sample from (e.g., np.random.exponential).
    - sample_size: The size of each sample drawn from the population.
    - num_samples: The number of samples to draw.
    - dist_params: Parameters for the population distribution (e.g., rate for exponential).

    Returns:
    - sample_means: List of means of each sample.
    """
    sample_means = []

    for _ in range(num_samples):
        sample = population_dist(*dist_params, size=sample_size)
        sample_means.append(np.mean(sample))

    return sample_means

# Parameters
population_dist = np.random.exponential  # Non-normal distribution (Exponential distribution)
dist_params = (1,)  # Rate (lambda) for the exponential distribution
sample_size = 30  # Size of each sample
num_samples = 1000  # Number of samples to draw

# Apply Central Limit Theorem
sample_means = central_limit_theorem(population_dist, sample_size, num_samples, dist_params)

# Plot the distribution of sample means
plt.figure(figsize=(8, 6))
plt.hist(sample_means, bins=30, edgecolor='black', alpha=0.7, density=True)
plt.xlabel('Sample Means')
plt.ylabel('Density')
plt.title('Central Limit Theorem: Sample Means Distribution')
plt.grid(True)
plt.show()
```

##Q19) Implement hypothesis testing using Z-statistics for a sample dataset .

Ans:

```python
import numpy as np
import scipy.stats as stats

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a Z-test for a sample dataset to test if the sample mean is significantly different from
    the population mean.

    Parameters:
    - sample_data: A list or numpy array of sample data.
    - population_mean: The mean of the population (null hypothesis value).
    - population_std: The population standard deviation.
    - alpha: Significance level (default 0.05).

    Returns:
    - z_statistic: The calculated Z-statistic.
    - p_value: The p-value for the test.
    - decision: The decision whether to reject the null hypothesis.
    """
    # Calculate sample mean and sample size
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Calculate the standard error of the mean (SEM)
    sem = population_std / np.sqrt(sample_size)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / sem

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    # Compare the p-value with the significance level to make a decision
    if p_value < alpha:
        decision = "Reject the null hypothesis"
    else:
        decision = "Fail to reject the null hypothesis"

    return z_statistic, p_value, decision

# Example: Sample data
sample_data = np.array([52, 55, 49, 60, 48, 50, 58, 53, 51, 47])

# Population parameters
population_mean = 50
population_std = 10

# Perform Z-test
z_stat, p_val, decision = z_test(sample_data, population_mean, population_std)

# Print results
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_val:.4f}")
print(f"Decision: {decision}")
```

##Q20)Create a confidence interval for a dataset using Python and interpret the result ?

Ans:

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculate the confidence interval for a dataset.

    Parameters:
    - sample_data: A list or numpy array of sample data.
    - confidence_level: The confidence level for the interval (default 0.95).

    Returns:
    - confidence_interval: The calculated confidence interval (lower bound, upper bound).
    """
    # Sample size
    n = len(sample_data)

    # Sample mean and sample standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Using ddof=1 for sample standard deviation

    # Degrees of freedom
    df = n - 1

    # t-critical value for the given confidence level
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df)

    # Standard error of the mean (SEM)
    sem = sample_std / np.sqrt(n)

    # Margin of error
    margin_of_error = t_critical * sem

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example: Sample data
sample_data = np.array([52, 55, 49, 60, 48, 50, 58, 53, 51, 47])

# Calculate the confidence interval (95% confidence level)
lower, upper = confidence_interval(sample_data, confidence_level=0.95)

# Print the result
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_normal_pdf(mean, std_dev):
    """
    Function to plot the Probability Density Function (PDF) of a normal distribution.

    Parameters:
    - mean: Mean of the normal distribution.
    - std_dev: Standard deviation of the normal distribution.
    """
    # Generate a range of x values from mean - 4*std_dev to mean + 4*std_dev
    x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

    # Calculate the PDF of the normal distribution for each x value
    pdf = stats.norm.pdf(x, loc=mean, scale=std_dev)

    # Plot the PDF
    plt.figure(figsize=(8, 6))
    plt.plot(x, pdf, label=f'Normal Distribution (mean={mean}, std={std_dev})', color='blue')
    plt.title('Probability Density Function (PDF) of a Normal Distribution')
    plt.xlabel('x')
    plt.ylabel('PDF')
    plt.grid(True)
    plt.legend()
    plt.show()

# Parameters for the normal distribution
mean = 0
std_dev = 1

# Plot the PDF
plot_normal_pdf(mean, std_dev)
```

##Q23)Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

Ans:


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_poisson_cdf(lmbda, max_value):
    """
    Function to calculate and plot the Cumulative Distribution Function (CDF) of a Poisson distribution.

    Parameters:
    - lmbda: The rate parameter (mean number of occurrences).
    - max_value: The maximum value for which to calculate the CDF.
    """
    # Generate x values from 0 to max_value
    x = np.arange(0, max_value+1)

    # Calculate the CDF for the Poisson distribution
    cdf = stats.poisson.cdf(x, lmbda)

    # Plot the CDF
    plt.figure(figsize=(8, 6))
    plt.step(x, cdf, label=f'Poisson CDF (Î»={lmbda})', where='post', color='green')
    plt.title('Cumulative Distribution Function (CDF) of a Poisson Distribution')
    plt.xlabel('k (Number of events)')
    plt.ylabel('CDF (P(X â‰¤ k))')
    plt.grid(True)
    plt.legend()
    plt.show()

# Parameters for the Poisson distribution
lmbda = 4  # average rate (mean) of occurrence
max_value = 15  # maximum value to calculate CDF for

# Plot the CDF for the Poisson distribution
plot_poisson_cdf(lmbda, max_value)
```

##Q24)Simulate a random variable using a continuous uniform distribution and calculate its expected value.

Ans:

```python
import numpy as np

# Function to simulate random variable and calculate expected value
def simulate_uniform_distribution(a, b, size=1000):
    """
    Simulate a random variable using a continuous uniform distribution and calculate its expected value.

    Parameters:
    - a: The lower bound of the distribution.
    - b: The upper bound of the distribution.
    - size: The number of random variables to simulate (default is 1000).

    Returns:
    - expected_value: The expected value calculated from the theoretical formula.
    - sample_mean: The mean of the simulated sample.
    """
    # Generate random variables from the continuous uniform distribution
    sample_data = np.random.uniform(a, b, size)

    # Calculate the expected value using the formula for a uniform distribution
    expected_value = (a + b) / 2

    # Calculate the sample mean from the simulated data
    sample_mean = np.mean(sample_data)

    return expected_value, sample_mean

# Parameters for the uniform distribution
a = 2  # lower bound
b = 8  # upper bound

# Simulate the random variable and calculate expected value
expected_value, sample_mean = simulate_uniform_distribution(a, b)

# Print results
print(f"Expected value (theoretical): {expected_value}")
print(f"Sample mean (calculated from simulation): {sample_mean}")
```

##Q25)Write a Python program to compare the standard deviations of two datasets and visualize the difference.

Ans:

```python
import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(data1, data2):
    """
    Compare the standard deviations of two datasets and visualize the difference.

    Parameters:
    - data1: First dataset (numpy array or list).
    - data2: Second dataset (numpy array or list).
    """
    # Calculate the standard deviations of both datasets
    std_dev_1 = np.std(data1)
    std_dev_2 = np.std(data2)

    print(f"Standard Deviation of Dataset 1: {std_dev_1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std_dev_2:.2f}")

    # Visualize the datasets using histograms
    plt.figure(figsize=(10, 6))

    # Plot histogram for Dataset 1
    plt.hist(data1, bins=15, alpha=0.5, label="Dataset 1", color="blue")

    # Plot histogram for Dataset 2
    plt.hist(data2, bins=15, alpha=0.5, label="Dataset 2", color="orange")

    plt.title("Comparison of Two Datasets")
    plt.xlabel("Data Values")
    plt.ylabel("Frequency")
    plt.legend(loc="upper right")
    plt.grid(True)
    plt.show()

# Generate two datasets with different standard deviations
np.random.seed(42)  # For reproducibility

# Dataset 1: Normal distribution with mean 50 and standard deviation 10
dataset1 = np.random.normal(loc=50, scale=10, size=1000)

# Dataset 2: Normal distribution with mean 50 and standard deviation 20
dataset2 = np.random.normal(loc=50, scale=20, size=1000)

# Compare the standard deviations of the two datasets and visualize
compare_standard_deviations(dataset1, dataset2)
```

##Q26)Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

Ans:

```python
import numpy as np

def calculate_range_and_iqr(data):
    """
    Calculate the range and interquartile range (IQR) of a given dataset.

    Parameters:
    - data: The dataset (numpy array or list).

    Returns:
    - range_value: The range of the dataset.
    - iqr: The interquartile range (IQR) of the dataset.
    """
    # Calculate the range (max - min)
    range_value = np.max(data) - np.min(data)

    # Calculate the interquartile range (IQR)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1

    return range_value, iqr

# Generate a dataset from a normal distribution
np.random.seed(42)  # For reproducibility
mean = 50
std_dev = 10
size = 1000

# Generate the dataset
data = np.random.normal(loc=mean, scale=std_dev, size=size)

# Calculate the range and IQR
range_value, iqr = calculate_range_and_iqr(data)

# Print results
print(f"Range of the dataset: {range_value:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {iqr:.2f}")
```

##Q27)Implement Z-score normalization on a dataset and visualize its transformation.

Ans:

```python
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Apply Z-score normalization to the given dataset.

    Parameters:
    - data: The dataset (numpy array or list).

    Returns:
    - normalized_data: The Z-score normalized dataset.
    """
    mean = np.mean(data)
    std_dev = np.std(data)

    # Apply Z-score formula
    normalized_data = (data - mean) / std_dev

    return normalized_data

# Generate a dataset from a normal distribution
np.random.seed(42)  # For reproducibility
mean = 50
std_dev = 10
size = 1000

# Generate the dataset
data = np.random.normal(loc=mean, scale=std_dev, size=size)

# Apply Z-score normalization
normalized_data = z_score_normalization(data)

# Plot the original data and the normalized data
plt.figure(figsize=(12, 6))

# Histogram of original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='blue', alpha=0.7)
plt.title("Original Data (Normal Distribution)")
plt.xlabel("Data Values")
plt.ylabel("Frequency")

# Histogram of normalized data
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='green', alpha=0.7)
plt.title("Z-score Normalized Data")
plt.xlabel("Normalized Data Values")
plt.ylabel("Frequency")

# Show the plots
plt.tight_layout()
plt.show()

# Print the mean and standard deviation of the normalized data
print(f"Mean of the normalized data: {np.mean(normalized_data):.2f}")
print(f"Standard deviation of the normalized data: {np.std(normalized_data):.2f}")
```

##Q28)Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal  distribution.

Ans:

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_and_kurtosis(data):
    """
    Calculate the skewness and kurtosis of a given dataset.

    Parameters:
    - data: The dataset (numpy array or list).

    Returns:
    - skew_value: The skewness of the dataset.
    - kurtosis_value: The kurtosis of the dataset.
    """
    # Calculate skewness
    skew_value = skew(data)

    # Calculate kurtosis (default is Fisher's kurtosis, where normal = 0)
    kurtosis_value = kurtosis(data)

    return skew_value, kurtosis_value

# Generate a dataset from a normal distribution
np.random.seed(42)  # For reproducibility
mean = 50
std_dev = 10
size = 1000

# Generate the dataset
data = np.random.normal(loc=mean, scale=std_dev, size=size)

# Calculate skewness and kurtosis
skew_value, kurtosis_value = calculate_skewness_and_kurtosis(data)

# Print results
print(f"Skewness of the dataset: {skew_value:.4f}")
print(f"Kurtosis of the dataset: {kurtosis_value:.4f}")
```
