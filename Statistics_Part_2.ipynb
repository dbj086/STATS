##Q1) What is hypothesis testing in statistics ?

Ans: **Hypothesis testing** in statistics is a method used to make decisions or inferences about population parameters based on sample data. It involves testing an assumption (called a **hypothesis**) about a population parameter.

### Key Concepts in Hypothesis Testing:

1. **Null Hypothesis (H‚ÇÄ)**:
   This is the default assumption or status quo. It usually states that there is **no effect**, **no difference**, or **no change**.

   * Example: H‚ÇÄ: The average height of adult men is 70 inches.

2. **Alternative Hypothesis (H‚ÇÅ or Ha)**:
   This is what you aim to support; it represents a new claim or an effect.

   * Example: Ha: The average height of adult men is **not** 70 inches.

3. **Test Statistic**:
   A value calculated from the sample data that is used to decide whether to reject the null hypothesis.

   * Common test statistics include *z*, *t*, *chi-square*, and *F*.

4. **Significance Level (Œ±)**:
   A threshold probability used to determine when to reject H‚ÇÄ. Common values are 0.05, 0.01, or 0.10.

   * If the probability of observing the test statistic under H‚ÇÄ is less than Œ±, we reject H‚ÇÄ.

5. **p-value**:
   The probability of getting a result at least as extreme as the one observed, assuming H‚ÇÄ is true.

   * If the p-value < Œ±, we reject H‚ÇÄ.

6. **Conclusion**:
   Based on the p-value and Œ±:

   * If p-value < Œ± ‚Üí **Reject H‚ÇÄ** (evidence supports Ha)
   * If p-value ‚â• Œ± ‚Üí **Fail to reject H‚ÇÄ** (not enough evidence to support Ha)

The **significance level**, denoted by **Œ± (alpha)**, is a threshold used to decide whether to reject the **null hypothesis (H‚ÇÄ)**. It represents the **probability of making a Type I error**, which is **rejecting H‚ÇÄ when it is actually true**.

##Q4)What does a P-value represent in hypothesis testing ?

Ans:### What Is a **P-value** in Hypothesis Testing?

The **p-value** (short for *probability value*) is a key result in hypothesis testing. It tells you **how likely your sample results would be if the null hypothesis (H‚ÇÄ) were true**.

---

### üîç In Simple Terms:

> The p-value is the probability of getting a result **at least as extreme** as the one you observed, **assuming H‚ÇÄ is true**.

---

### üìä What Does the P-value Tell You?

* **Small p-value (e.g., < 0.05)**:

  * Your observed data is **unlikely** under the null hypothesis.
  * You have **evidence to reject H‚ÇÄ** in favor of the alternative hypothesis (Ha).

* **Large p-value (e.g., ‚â• 0.05)**:

  * Your observed data is **likely** under H‚ÇÄ.
  * You **do not reject H‚ÇÄ**; there isn‚Äôt strong enough evidence for Ha.

---

### üß† Example:

You‚Äôre testing whether a coin is fair.

* **H‚ÇÄ**: The coin is fair (P = 0.5).
* You flip the coin 100 times and get 70 heads.
* A statistical test gives you a **p-value of 0.003**.

#### Interpretation:

* There‚Äôs only a **0.3% chance** you would see 70 or more heads **if the coin were fair**.
* Since 0.003 < 0.05, you **reject H‚ÇÄ**. There‚Äôs strong evidence the coin is not fair.

---

### üö¶ How to Use the P-value:

| **P-value** | **Conclusion (if Œ± = 0.05)**                      |
| ----------- | ------------------------------------------------- |
| p ‚â§ 0.05    | Reject H‚ÇÄ ‚Üí statistically significant result      |
| p > 0.05    | Fail to reject H‚ÇÄ ‚Üí not statistically significant |

> **Important**: A small p-value does **not** prove H‚ÇÄ is false or that Ha is true. It just means the data would be **very unusual** if H‚ÇÄ were true.

##Q6)What are Type 1 and Type 2 errors in hypothesis testing ?

Ans:Great question! Understanding **Type I and Type II errors** is essential in hypothesis testing because they explain the risks involved in making decisions based on data.

---

### ‚úÖ **Overview of Hypothesis Testing Outcomes**

In hypothesis testing, there are two possible decisions:

* **Reject the null hypothesis (H‚ÇÄ)**
* **Fail to reject the null hypothesis (H‚ÇÄ)**

And there are two realities:

* **H‚ÇÄ is actually true**
* **H‚ÇÄ is actually false**

Combining these gives **four possible outcomes**:

| **Decision**          | **H‚ÇÄ True**           | **H‚ÇÄ False**                |
| --------------------- | --------------------- | --------------------------- |
| **Reject H‚ÇÄ**         | ‚ùå **Type I Error**    | ‚úÖ Correct (Detected Effect) |
| **Fail to Reject H‚ÇÄ** | ‚úÖ Correct (No Effect) | ‚ùå **Type II Error**         |

---

### üî¥ **Type I Error (False Positive)**

* **Definition**: Rejecting H‚ÇÄ when it is actually true.
* **You falsely conclude** there is an effect or difference.
* **Probability** of Type I Error = **Œ± (significance level)**

  * Common Œ± values: 0.05, 0.01

> **Example**: You conclude a drug works when it actually doesn‚Äôt.

---

### üîµ **Type II Error (False Negative)**

* **Definition**: Failing to reject H‚ÇÄ when it is actually false.
* **You miss a real effect** that does exist.
* **Probability** of Type II Error = **Œ≤**

  * Related to **power = 1 ‚àí Œ≤**, the ability to detect a real effect.

> **Example**: You conclude a drug doesn't work when it actually does.

---

### üéØ **Balancing the Errors**

* Lowering Œ± reduces the chance of a **Type I error** but increases the risk of a **Type II error**.
* Increasing the **sample size** can reduce both errors.

---

### üîÅ Real-World Analogy: Courtroom

| Legal Decision | Reality: Innocent    | Reality: Guilty            |
| -------------- | -------------------- | -------------------------- |
| **Convict**    | ‚ùå Type I Error       | ‚úÖ Correct (justice served) |

> Convicting an innocent = Type I
> Letting a criminal go free = Type II

##Q8) What is the Z-test, and when is it used in hypothesis testing ?

Ans: ### ‚úÖ What is a **Z-test**?

The **Z-test** is a type of **statistical hypothesis test** used to determine whether there is a significant difference between a sample statistic and a known population parameter ‚Äî **when the population variance is known** (or the sample size is large).

It uses the **Z-distribution** (standard normal distribution), which has a mean of 0 and a standard deviation of 1.

---

### üîç When is the Z-test Used?

Use a **Z-test** when **ALL** the following conditions are met:

| Condition                                                                                           | Explanation                                            |
| --------------------------------------------------------------------------------------------------- | ------------------------------------------------------ |
| ‚úÖ The sample size is **large**                                                                      | Typically **n ‚â• 30**                                   |
| ‚úÖ The population **standard deviation (œÉ)** is known                                                | Unlike the **t-test**, which is used when œÉ is unknown |
| ‚úÖ Data is **normally distributed** (or sample size large enough for Central Limit Theorem to apply) | Ensures valid Z-test results                           |

---

### üß™ Types of Z-tests:

1. **One-Sample Z-test**:
   Tests whether the sample mean differs from a known population mean.

   * Example: Is the average SAT score of a class different from the national average of 1050?

2. **Two-Sample Z-test**:
   Compares the means of two independent groups.

   * Example: Is there a difference in average test scores between two schools?

3. **Z-test for Proportions**:
   Tests if a sample proportion differs from a population proportion (or between two proportions).

   * Example: Is the proportion of voters supporting a candidate different from 50%?

---

### üìê Z-test Formula (One-Sample Mean):

$$
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
$$

Where:

* $\bar{X}$ = sample mean
* $\mu$ = population mean
* $\sigma$ = population standard deviation
* $n$ = sample size

---

### üéØ Example:

Suppose:

* Population mean (Œº) = 100
* Population standard deviation (œÉ) = 15
* Sample mean (ùëãÃÑ) = 108
* Sample size (n) = 36

$$
Z = \frac{108 - 100}{15 / \sqrt{36}} = \frac{8}{2.5} = 3.2
$$

A Z-score of 3.2 is far in the tail ‚Äî leading to a very low p-value ‚Üí **Reject H‚ÇÄ**.

##Q10) What is the T-distribution, and when should it be used instead of the normal distribution ?

| Feature                            | Description                                                    |
| ---------------------------------- | -------------------------------------------------------------- |

### üß™ When to Use the T-distribution Instead of the Normal (Z) Distribution:

| Use **T-distribution** when...                             |
| ---------------------------------------------------------- |
| ‚úÖ The **population standard deviation (œÉ)** is **unknown** |
| ‚úÖ The **sample size is small** (typically n < 30)          |

> **If œÉ is known** and sample size is large ‚Üí use the **Z-distribution**
> **If œÉ is unknown** ‚Üí use the **T-distribution**

---

### üìê Formula for the T-statistic (One-Sample Test):

$$

$$

Where:

* $\bar{X}$ = sample mean

##Q12)What is the T-test, and how is it used in hypothesis testing ?

Ans:### What Is a **T-test**?

A **T-test** is a statistical method used to determine if there is a significant difference between:

* The mean of a sample and a known value (usually the population mean), or
* The means of two groups (independent or paired samples),

when the **population standard deviation (œÉ) is unknown** and the sample size is relatively **small**.

---

### Why Use a T-test?

* When **œÉ is unknown**, you estimate variability using the **sample standard deviation (s)**.
* The T-test accounts for this uncertainty using the **T-distribution**, which has heavier tails than the normal distribution.
* This helps avoid overestimating significance when the sample is small.

---

### Common Types of T-tests

| Test Type                         | Use Case                                                               |
| --------------------------------- | ---------------------------------------------------------------------- |
| **One-Sample T-test**             | Compare sample mean to a known population mean.                        |
| **Independent Two-Sample T-test** | Compare means of two independent groups (e.g., drug vs. placebo).      |
| **Paired Sample T-test**          | Compare means from the same group at two time points or matched pairs. |

---

### How the T-test Works (One-Sample Example)

1. **Set hypotheses:**

$$
H_0: \mu = \mu_0 \quad \text{(no difference)}
$$

$$
H_a: \mu \neq \mu_0 \quad \text{(two-tailed)}
$$

2. **Calculate the T-statistic:**

$$
t = \frac{\bar{X} - \mu_0}{s / \sqrt{n}}
$$

Where:

* $\bar{X}$ = sample mean
* $\mu_0$ = population mean under H‚ÇÄ
* $s$ = sample standard deviation
* $n$ = sample size

3. **Determine degrees of freedom (df):**

$$
df = n - 1
$$

4. **Find the critical t-value or p-value** from the t-distribution table using $df$.

5. **Make a decision:**

* If $|t| > \text{critical value}$ or p-value < significance level (Œ±), **reject H‚ÇÄ**.
* Otherwise, **fail to reject H‚ÇÄ**.

---

### Example:

* Sample size $n = 20$
* Sample mean $\bar{X} = 105$
* Population mean $\mu_0 = 100$
* Sample standard deviation $s = 15$
* Significance level $\alpha = 0.05$

Calculate:

$$
t = \frac{105 - 100}{15 / \sqrt{20}} = \frac{5}{3.35} \approx 1.49
$$

Degrees of freedom = 19.

Check the t-table for $df = 19$ at Œ±=0.05 (two-tailed), critical value ‚âà 2.093.

Since $1.49 < 2.093$, **fail to reject H‚ÇÄ** ‚Äî no significant difference.

* When **œÉ is unknown** and the **sample size is small**, the **T-test** is better because it accounts for the added uncertainty in estimating œÉ using the sample standard deviation. The heavier tails of the t-distribution reflect this uncertainty.
* As the **sample size grows larger**, the **t-distribution approaches the normal distribution**, making the **T-test and Z-test results nearly the same**.

##Q14) What is a confidence interval, and how is it used to interpret statistical results ?

Ans: ### What Is a **Confidence Interval (CI)**?

A **confidence interval** is a **range of values** calculated from sample data that is likely to contain the **true population parameter** (like the population mean) with a specified level of confidence.

---

### Key Points:

* It gives an **estimate range** rather than a single point estimate.
* The confidence level (e.g., 95%) tells us how confident we are that the interval contains the true parameter.
* It reflects the **uncertainty** due to sampling variability.

---

### Example:

A **95% confidence interval** means:

> If you took many samples and calculated a confidence interval from each, **about 95% of those intervals would contain the true population parameter**.

---

### How to Calculate a Confidence Interval for a Mean:

$$
\text{CI} = \bar{X} \pm (\text{critical value}) \times \frac{s}{\sqrt{n}}
$$

Where:

* $\bar{X}$ = sample mean
* $s$ = sample standard deviation (or population SD if known)
* $n$ = sample size
* **Critical value** = depends on confidence level and distribution (Z for large samples or known œÉ, t for small samples and unknown œÉ)

---

### Example:

* Sample mean $\bar{X} = 100$
* Sample standard deviation $s = 15$
* Sample size $n = 25$
* Confidence level = 95%
* Degrees of freedom = 24
* Critical t-value ‚âà 2.064

Calculate margin of error:

$$
ME = 2.064 \times \frac{15}{\sqrt{25}} = 2.064 \times 3 = 6.192
$$

So, the 95% CI is:

$$
100 \pm 6.192 = (93.81, 106.19)
$$

---

### How Is It Used in Interpreting Results?

* **If the CI for a mean difference includes 0**, it suggests **no significant difference** at the chosen confidence level.
* **If the CI does NOT include the null hypothesis value**, it supports **rejecting the null hypothesis**.
* Gives a **range of plausible values** for the population parameter, helping understand precision.

---

### Why Confidence Intervals Are Useful:

* They provide more information than just a p-value.
* They show **both the estimate and the uncertainty**.
* Help assess the **practical significance** of results.

##Q15) What is the margin of error, and how does it affect the confidence interval ?

ANS:The **margin of error** and the **confidence interval** are closely related statistical concepts used when estimating a population parameter (like a mean or proportion) based on a sample.

---

### üîπ **What is the Margin of Error?**

The **margin of error (MoE)** is a value that shows the range within which the true population parameter is likely to fall, given a certain level of confidence.

It accounts for the variability in sampling and is usually expressed as:

$$
\text{Margin of Error} = z^* \times \text{Standard Error (SE)}
$$

Where:

* $z^*$ is the **critical value** from the standard normal distribution (based on the chosen confidence level),
* **Standard Error (SE)** measures the spread of the sampling distribution.

---

### üîπ **How It Affects the Confidence Interval**

The **confidence interval (CI)** gives a range of values believed to contain the true population parameter. It is calculated as:

$$
\text{Confidence Interval} = \text{Sample Estimate} \pm \text{Margin of Error}
$$

So, the **margin of error determines the width** of the confidence interval:

* **Larger MoE ‚Üí Wider CI** (more uncertainty)
* **Smaller MoE ‚Üí Narrower CI** (more precision)

---

### üîπ Example

Suppose you're estimating the average height of adult males in a city:

* Sample mean = 70 inches
* Margin of error = 2 inches
* Confidence level = 95%

Then your confidence interval is:

$$
70 \pm 2 \rightarrow [68, 72] \text{ inches}
$$

You can say: *‚ÄúWe are 95% confident that the true average height is between 68 and 72 inches.‚Äù*

---

### üîπ What Affects the Margin of Error?

* **Sample Size (n):** Larger sample ‚Üí smaller SE ‚Üí smaller MoE
* **Confidence Level:** Higher confidence (e.g., 99% vs 95%) ‚Üí larger $z^*$ ‚Üí larger MoE
* **Population variability:** More variability ‚Üí larger SE ‚Üí larger MoE

##Q16) How is Bayes' Theorem used in statistics, and what is its significance ?

---

### üîπ **What is Bayes‚Äô Theorem?**

Bayes‚Äô Theorem is stated mathematically as:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

Where:

* $P(A|B)$ is the **posterior probability**: the probability of hypothesis $A$ given observed data $B$
* $P(B|A)$ is the **likelihood**: the probability of observing data $B$ given that hypothesis $A$ is true
* $P(A)$ is the **prior probability**: your belief in hypothesis $A$ before seeing the data
* $P(B)$ is the **marginal probability**: the total probability of observing the data under all hypotheses

---

### üîπ **How It‚Äôs Used in Statistics**

#### ‚úÖ 1. **Updating Beliefs**

Bayes‚Äô Theorem allows you to revise probabilities as more data becomes available. This is the core idea of **Bayesian inference**.

#### ‚úÖ 2. **Parameter Estimation**

In Bayesian statistics, unknown parameters are treated as random variables. Bayes‚Äô Theorem is used to compute their **posterior distributions**.

#### ‚úÖ 3. **Model Comparison**

Bayesian methods can compare models using **Bayes factors**, which assess how well each model explains the observed data.

#### ‚úÖ 4. **Classification and Machine Learning**

Used in algorithms like the **Naive Bayes classifier**, which applies Bayes‚Äô Theorem under independence assumptions to classify data efficiently.

---

### üîπ **Why It‚Äôs Significant**

| Feature                          | Significance                                                              |
| -------------------------------- | ------------------------------------------------------------------------- |
| **Intuitive Updating**           | Mimics rational thinking: adjust your beliefs as new data comes in        |
| **Incorporates Prior Knowledge** | Allows experts to include prior experience or historical data             |
| **Probabilistic Interpretation** | Gives full probability distributions, not just point estimates            |

##Q1) Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and  interpret the results ?

Ans:

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Performs a two-tailed Z-test comparing sample mean to population mean.

    Parameters:
    sample (list or array): Sample data
    pop_mean (float): Known population mean
    pop_std (float): Known population standard deviation
    alpha (float): Significance level (default 0.05)

    Returns:
    z_score, p_value, interpretation
    """
    sample_mean = np.mean(sample)
    n = len(sample)
    standard_error = pop_std / np.sqrt(n)

    # Calculate Z-score
    z_score = (sample_mean - pop_mean) / standard_error

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Interpretation
    if p_value < alpha:
        interpretation = "Reject the null hypothesis: significant difference."
    else:
        interpretation = "Fail to reject the null hypothesis: no significant difference."

    return z_score, p_value, interpretation


# Example data
sample_data = [105, 110, 98, 95, 102, 100, 108, 112, 99, 101]
population_mean = 100
population_std_dev = 15
significance_level = 0.05

# Perform Z-test
z, p, result = z_test(sample_data, population_mean, population_std_dev, significance_level)

print(f"Sample mean = {np.mean(sample_data):.2f}")
print(f"Z-score = {z:.3f}")
print(f"P-value = {p:.4f}")
print(f"Result: {result}")
```

##Q5) Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing ?

Ans:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05):
    """
    Calculate and visualize Type I and Type II errors for testing H0: mu=mu0 vs Ha: mu=mu1.

    Parameters:
    - mu0: mean under null hypothesis
    - mu1: mean under alternative hypothesis
    - sigma: population standard deviation (known)
    - n: sample size
    - alpha: significance level (Type I error rate)
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for two-tailed test
    z_crit_low = norm.ppf(alpha / 2)
    z_crit_high = norm.ppf(1 - alpha / 2)

    # Critical values on original scale
    crit_low = mu0 + z_crit_low * se
    crit_high = mu0 + z_crit_high * se

    # Calculate Type II error (beta):
    # Probability of failing to reject H0 when Ha is true,
    # i.e., sample mean falls between critical values under Ha distribution
    beta = norm.cdf(crit_high, loc=mu1, scale=se) - norm.cdf(crit_low, loc=mu1, scale=se)

    # Plotting x range covering both distributions well
    x_min = min(mu0, mu1) - 4*se
    x_max = max(mu0, mu1) + 4*se
    x = np.linspace(x_min, x_max, 1000)

    # PDFs
    pdf_H0 = norm.pdf(x, mu0, se)
    pdf_Ha = norm.pdf(x, mu1, se)

    plt.figure(figsize=(12,6))

    # Plot distributions
    plt.plot(x, pdf_H0, label='Null Hypothesis H0', color='blue')
    plt.plot(x, pdf_Ha, label='Alternative Hypothesis Ha', color='green')

    # Shade Type I error regions (tails under H0)
    plt.fill_between(x, 0, pdf_H0, where=(x <= crit_low), color='red', alpha=0.3, label='Type I Error (Œ±)')
    plt.fill_between(x, 0, pdf_H0, where=(x >= crit_high), color='red', alpha=0.3)

    # Shade Type II error region (middle area under Ha)
    plt.fill_between(x, 0, pdf_Ha, where=((x > crit_low) & (x < crit_high)), color='orange', alpha=0.3, label='Type II Error (Œ≤)')

    # Critical value lines
    plt.axvline(crit_low, color='black', linestyle='--', label='Critical Values')
    plt.axvline(crit_high, color='black', linestyle='--')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Type I error (Œ±): {alpha:.3f}")
    print(f"Type II error (Œ≤): {beta:.3f}")
    print(f"Power of test (1 - Œ≤): {1 - beta:.3f}")

# Example usage:
plot_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05)
```

##Q9) Write a Python function to calculate the confidence interval for a sample mean and explain its significance ?

Ans:

```python
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
    - sample: list or numpy array of sample data
    - confidence: confidence level (default 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound): tuple representing the confidence interval
    """
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # Standard error of the mean

    # Degrees of freedom
    df = n - 1

    # t critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df)

    margin_of_error = t_crit * std_err

    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
sample_data = [12, 15, 14, 16, 13, 15, 14, 16, 15, 14]
conf_level = 0.95

lower, upper = confidence_interval(sample_data, conf_level)
print(f"{conf_level*100:.0f}% confidence interval: ({lower:.2f}, {upper:.2f})")
```

##Q10) Write a Python program to calculate the margin of error for a given confidence level using sample data ?

```python
import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for the sample mean.

    Parameters:
    - sample: list or numpy array of sample data
    - confidence: confidence level (default 0.95)

    Returns:
    - margin_of_error: the margin of error
    """
    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # Standard error

    df = n - 1  # Degrees of freedom
    t_crit = t.ppf((1 + confidence) / 2, df)  # t critical value

    return t_crit * std_err

# Example usage
sample_data = [23, 21, 19, 25, 22, 20, 24, 23, 21, 22]
confidence_level = 0.95

moe = margin_of_error(sample_data, confidence_level)
print(f"Margin of Error at {confidence_level*100:.0f}% confidence: {moe:.3f}")
```

##Q11) Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process ?

Ans:

```python
def bayes_theorem(prior, likelihood, false_positive_rate):
    """
    Calculate the posterior probability using Bayes' theorem.

    Parameters:
    - prior: P(H), prior probability of the hypothesis (disease)
    - likelihood: P(E|H), probability of evidence if hypothesis is true (sensitivity)
    - false_positive_rate: P(E|~H), probability of evidence if hypothesis is false

    Returns:
    - posterior: P(H|E), updated probability of hypothesis given evidence
    """
    # Probability of evidence P(E)
    prob_evidence = (likelihood * prior) + (false_positive_rate * (1 - prior))

    # Posterior probability P(H|E)
    posterior = (likelihood * prior) / prob_evidence

    return posterior

# Given values
prior_disease = 0.01
sensitivity = 0.99
false_positive_rate = 0.05

posterior_prob = bayes_theorem(prior_disease, sensitivity, false_positive_rate)
print(f"Posterior probability of having the disease given a positive test: {posterior_prob:.4f}")
```

##Q13) Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data ?  

Ans:

```python
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test.

    Parameters:
    - observed: 2D numpy array of observed counts

    Returns:
    - expected: 2D numpy array of expected counts
    """
    # Row totals and column totals
    row_totals = observed.sum(axis=1).reshape(-1, 1)
    col_totals = observed.sum(axis=0).reshape(1, -1)
    total = observed.sum()

    # Expected frequency = (row total * column total) / grand total
    expected = (row_totals @ col_totals) / total
    return expected

# Example observed data
observed = np.array([[30, 10],
                     [20, 40]])

expected = calculate_expected_frequencies(observed)
print("Observed Frequencies:\n", observed)
print("Expected Frequencies:\n", expected)
```

##Q16) Implement an F-test using Python to compare the variances of two random samples ?

Ans:


```python
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two samples.

    Parameters:
    - sample1, sample2: arrays or lists of sample data
    - alpha: significance level (default 0.05)

    Returns:
    - F_stat: computed F statistic
    - p_value: two-tailed p-value
    - conclusion: string interpreting the result
    """
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Calculate F statistic (ratio of larger variance to smaller variance)
    if var1 > var2:
        F_stat = var1 / var2
        dfn = len(sample1) - 1  # degrees of freedom numerator
        dfd = len(sample2) - 1  # degrees of freedom denominator
    else:
        F_stat = var2 / var1
        dfn = len(sample2) - 1
        dfd = len(sample1) - 1

    # Two-tailed p-value calculation
    p_value = 2 * min(f.cdf(F_stat, dfn, dfd), 1 - f.cdf(F_stat, dfn, dfd))
    p_value = min(p_value, 1)  # p-value can't exceed 1

    conclusion = ("Reject null hypothesis: variances are significantly different."
                  if p_value < alpha else
                  "Fail to reject null hypothesis: no significant difference in variances.")

    return F_stat, p_value, conclusion

# Example samples
np.random.seed(42)
sample1 = np.random.normal(loc=10, scale=5, size=30)
sample2 = np.random.normal(loc=10, scale=7, size=30)

F_stat, p_value, conclusion = f_test(sample1, sample2)
print(f"F-statistic: {F_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(conclusion)
```

##Q19)Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA ?

Ans:

```python
import numpy as np
from scipy.stats import shapiro, levene

def check_anova_assumptions(groups, alpha=0.05):
    """
    Check ANOVA assumptions: normality and equal variances.
    Independence must be ensured by study design.

    Parameters:
    - groups: list of arrays/lists, each containing data for one group
    - alpha: significance level for tests (default 0.05)

    Returns:
    - dict with test results for normality and equal variances
    """
    results = {'normality': {}, 'equal_variance': None, 'independence': 'Assumed by study design'}

    # Normality test per group
    for i, group in enumerate(groups, start=1):
        stat, p = shapiro(group)
        results['normality'][f'Group {i}'] = {
            'statistic': stat,
            'p_value': p,
            'normal': p > alpha
        }

    # Equal variance test (Levene's test)
    stat, p = levene(*groups)
    results['equal_variance'] = {
        'statistic': stat,
        'p_value': p,
        'equal_variance': p > alpha
    }

    return results

# Example usage
group1 = [23, 21, 19, 25, 22]
group2 = [30, 29, 31, 28, 35]
group3 = [27, 26, 24, 28, 30]

assumption_results = check_anova_assumptions([group1, group2, group3])

print("ANOVA Assumptions Check:")
for group, result in assumption_results['normality'].items():
    print(f"{group} normality: p = {result['p_value']:.4f} -> {'Normal' if result['normal'] else 'Not Normal'}")

eq_var = assumption_results['equal_variance']
print(f"Equal variances (Levene's test): p = {eq_var['p_value']:.4f} -> {'Equal variances' if eq_var['equal_variance'] else 'Variances differ'}")
print(f"Independence: {assumption_results['independence']}")
```

##Q25)Write a Python script to perform a Z-test for comparing proportions between two datasets or groups ?

Ans:

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

def two_proportion_z_test(success_a, size_a, success_b, size_b, alpha=0.05):
    """
    Perform two-proportion Z-test.

    Parameters:
    - success_a: number of successes in group A
    - size_a: total observations in group A
    - success_b: number of successes in group B
    - size_b: total observations in group B
    - alpha: significance level (default 0.05)

    Returns:
    - z_stat: computed Z statistic
    - p_value: p-value of the test
    - conclusion: interpretation of result
    """
    count = np.array([success_a, success_b])
    nobs = np.array([size_a, size_b])

    z_stat, p_value = proportions_ztest(count, nobs)

    conclusion = ("Reject null hypothesis: proportions are significantly different."
                  if p_value < alpha else
                  "Fail to reject null hypothesis: no significant difference between proportions.")

    return z_stat, p_value, conclusion

# Example usage:
success_a = 45
size_a = 100

success_b = 30
size_b = 100

z_stat, p_value, conclusion = two_proportion_z_test(success_a, size_a, success_b, size_b)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(conclusion)
```
